```python
# =====================================
# TASK 2: Exploratory Data Analysis (EDA)
# Internship Ready Program
# =====================================

import pandas as pd
import matplotlib.pyplot as plt

# ---------------- STEP 1: Define Data ----------------
data = {
    'Age': [22, 25, 30, 35, 40],
    'Salary': [20000, 25000, 30000, 40000, 50000],
    'Experience': [1, 2, 5, 8, 12]
}

df = pd.DataFrame(data)

print("STEP 1: Dataset Loaded Successfully")
print(df)

# ---------------- STEP 2: Meaningful Questions ----------------
print("\nSTEP 2: Meaningful Questions")
print("1. What is the average salary?")
print("2. How does salary vary with age?")
print("3. What is the distribution of salary?")
print("4. How experience affects salary?")
print("5. What are the minimum and maximum values?")

# ---------------- STEP 3: Data Structure Exploration ----------------
print("\nSTEP 3: Data Structure")
print("Shape (Rows, Columns):", df.shape)
print("Column Names:", df.columns.tolist())
print("\nData Types:")
print(df.dtypes)

# ---------------- STEP 4: Summary Statistics ----------------
print("\nSTEP 4: Summary Statistics")
print(df.describe())

# ---------------- STEP 5: Data Visualization ----------------

# Salary Distribution - Histogram
plt.figure()
plt.hist(df['Salary'])
plt.xlabel("Salary")
plt.ylabel("Frequency")
plt.title("Salary Distribution")
plt.show()

# Age vs Salary - Scatter Plot
plt.figure()
plt.scatter(df['Age'], df['Salary'])
plt.xlabel("Age")
plt.ylabel("Salary")
plt.title("Age vs Salary")
plt.show()

# Experience vs Salary - Line Chart
plt.figure()
plt.plot(df['Experience'], df['Salary'])
plt.xlabel("Experience (Years)")
plt.ylabel("Salary")
plt.title("Experience vs Salary")
plt.show()

print("\nEDA Task 2 Completed Successfully ")
```
