```python
import requests
from bs4 import BeautifulSoup

# STEP 1: Identify the website (public website)
url = "https://quotes.toscrape.com/"

# STEP 2: Send request to the website
response = requests.get(url)

# STEP 3: Parse HTML content
soup = BeautifulSoup(response.text, "html.parser")

# STEP 4: Extract required data (quotes & authors)
quotes = soup.find_all("span", class_="text")
authors = soup.find_all("small", class_="author")

print("Scraped Quotes:\n")

# STEP 5: Display the extracted data
for i in range(len(quotes)):
    print(f"Quote {i+1}: {quotes[i].text}")
    print(f"Author: {authors[i].text}")
    print("-" * 50)

print("Task 1 Web Scraping completed successfully")
```
